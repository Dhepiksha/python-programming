```python

account_holders = {"C101", "C102", "C103", "C104", "C105", "C106"}
kyc_verified = {"C102", "C104", "C106", "C108"}
defaulters = {"C107", "C105"}

# 1️ Create set of customers eligible for loan
eligible_customers = account_holders & kyc_verified - defaulters

print("Eligible customers:", eligible_customers)

# 2️ Display total eligible customers
print("Total eligible customers:", len(eligible_customers))

# 3️ Check whether C104 is eligible
print("Is C104 eligible?", "C104" in eligible_customers)

# 4️ Customers either verified or defaulters but not both
result = kyc_verified ^ defaulters
print("Verified or Defaulters but not both:", result)

# 5️ Update operations
kyc_verified.add("C107") # add C107

account_holders.discard("C105") # remove safely

print("Updated KYC verified:", kyc_verified)
print("Updated account holders:", account_holders)

# 6️ Clean up
backup = kyc_verified.copy()

kyc_verified.clear()

print("Backup KYC:", backup)
print("Original KYC after clear:", kyc_verified)
```

```python
savings_customers = {"C101","C102","C103","C104","C105"}
loan_customers = {"C104","C105","C106","C107"}
credit_card_customers = {"C102","C105","C108"}

# 1. Unique customers using at least one service (Union)
all_customers = savings_customers | loan_customers | credit_card_customers
print("All customers:", all_customers)

# 2. Customers having both savings and loan
both_savings_loan = savings_customers & loan_customers
print("Savings and Loan:", both_savings_loan)

# 3. Customers with loan but not savings
loan_not_savings = loan_customers - savings_customers
print("Loan but not Savings:", loan_not_savings)

# 4. Customers with only savings or credit card (not both)
only_one = savings_customers ^ credit_card_customers
print("Savings or Credit card but not both:", only_one)

# 5. Check if credit card customers are subset of savings
print("Credit card subset of savings:", credit_card_customers.issubset(savings_customers))

# 6. Check if loan and credit card customers are disjoint
print("Loan and Credit card disjoint:", loan_customers.isdisjoint(credit_card_customers))

# 7. Check if savings customers include all loan customers (superset)
print("Savings superset of loan:", savings_customers.issuperset(loan_customers))

# 8. Remove credit card customers from savings (in-place update)
savings_customers.difference_update(credit_card_customers)
print("Savings after removal:", savings_customers)
```

```python
from functools import reduce

# Step 1: List of daily transactions
daily_transactions = [5000, 1500, -200, 12000, 300, -500, 700000]

print("All Transactions:")
print(daily_transactions)

# Step 2: Filter Deposits (only positive values)
deposits = list(filter(lambda x: x > 0, daily_transactions))

print("\nFiltered Deposits (Positive Transactions):")
print(deposits)

# Step 3: Map for Commission (0.05% commission)
commission_rates = list(map(lambda x: x * 0.0005, deposits))

print("\nCommission on each deposit (0.05%):")
print(commission_rates)

# Step 4: Reduce to calculate total commission
total_commission = reduce(lambda a, b: a + b, commission_rates)

print("\nTotal Commission Earned:")
print(total_commission)
```

```python
flights = {
    "AI202": 10,
    "6E501": 5
}

passengers = []

def show_status():
    print("\nCurrent Seat Availability")
    for name, count in flights.items():
        print("Flight", name, ":", count, "seats left")

    print("\nPassenger List")
    if not passengers:
        print("No passengers booked yet.")
    else:
        for p in passengers:
            print(p["name"], "-", p["flight"])


def book_ticket():
    name = input("Enter Passenger Name: ")
    age = int(input("Enter Age: "))
    phone = input("Enter Phone Number: ")
    f_id = input("Enter Flight ID (AI202 or 6E501): ")

    if f_id in flights:
        if flights[f_id] > 0:
            flights[f_id] -= 1

            passenger = {
                "name": name,
                "age": age,
                "phone": phone,
                "flight": f_id
            }

            passengers.append(passenger)
            print("Ticket booked successfully for", name)
        else:
            print("No seats available")
    else:
        print("Flight not found")


def cancel_ticket():
    name = input("Enter Passenger Name to cancel: ")

    for p in passengers:
        if p["name"] == name:
            flights[p["flight"]] += 1
            passengers.remove(p)
            print("Ticket cancelled for", name)
            return

    print("Passenger not found")


def add_seats():
    f_id = input("Enter Flight ID: ")
    seats = int(input("Enter number of seats to add: "))

    if f_id in flights:
        flights[f_id] += seats
        print(seats, "seats added to", f_id)
    else:
        print("Flight not found")


def generate_ticket():
    name = input("Enter Passenger Name: ")

    for p in passengers:
        if p["name"] == name:
            print("\n----- TICKET -----")
            print("Passenger Name:", p["name"])
            print("Age:", p["age"])
            print("Phone:", p["phone"])
            print("Flight ID:", p["flight"])
            print("------------------")
            return

    print("Passenger not found")


while True:
    print("\nAIRLINE SYSTEM")
    print("1 View Flights")
    print("2 Book Ticket")
    print("3 Cancel Ticket")
    print("4 Exit")
    print("5 Add Seat Availability")
    print("6 Generate Ticket")

    choice = input("Enter choice: ")

    if choice == "1":
        show_status()
    elif choice == "2":
        book_ticket()
    elif choice == "3":
        cancel_ticket()
    elif choice == "5":
        add_seats()
    elif choice == "6":
        generate_ticket()
    elif choice == "4":
        print("System Closed")
        break
    else:
        print("Invalid choice")
```

```python
import datetime
class User:
    def __init__(self, user_id, username, email):
        self.user_id = user_id
        self.username = username
        self.email = email
        self.posts = 0
        self.engagement = 0
    def add_post(self, engagement):
        self.posts += 1
        self.engagement += engagement
    def display_user(self):
        print("User ID:", self.user_id)
        print("Username:", self.username)
        print("Email:", self.email)
        print("Posts:", self.posts)
        print("Total Engagement:", self.engagement)

class SocialMediaAnalyzer:

    def __init__(self):
        self.users = {}
        self.posts = []
        self.hashtags = set()
        self.hashtag_frequency = {}
    def register_user(self, user_id, username, email):
        if user_id in self.users:
            print("User already exists")
            return
        user = User(user_id, username, email)
        self.users[user_id] = user
        print("User Registered Successfully")
    def view_users(self):
        if not self.users:
            print("No users available")
            return
        for user in self.users.values():
            user.display_user()
    def extract_hashtags(self, content):
        words = content.split()
        tags = []
        for word in words:
            if word.startswith("#"):
                tag = word[1:]
                tags.append(tag)
                self.hashtags.add(tag)

                if tag in self.hashtag_frequency:
                    self.hashtag_frequency[tag] += 1
                else:
                    self.hashtag_frequency[tag] = 1
        return tags
    def add_post(self, post_id, user_id, content, likes, comments):
        if user_id not in self.users:
            print("User not found")
            return
        hashtags = self.extract_hashtags(content)
        post = {
            "post_id": post_id,
            "user_id": user_id,
            "content": content,
            "hashtags": hashtags,
            "likes": likes,
            "comments": comments,
            "timestamp": datetime.datetime.now()
        }
        self.posts.append(post)
        engagement = likes + comments
        self.users[user_id].add_post(engagement)
        print("Post added successfully")
    def trending_hashtags(self):

        print("\nTrending Hashtags")

        sorted_tags = sorted(
            self.hashtag_frequency.items(),
            key=lambda x: x[1],
            reverse=True
        )

        for tag, count in sorted_tags:
            print("#", tag, "-", count, "posts")
    def most_active_users(self):
        print("\nMost Active Users")
        sorted_users = sorted(
            self.users.values(),
            key=lambda x: (x.posts, x.engagement),
            reverse=True
        )
        for user in sorted_users:
            print(
                "User:", user.username,
                "| Posts:", user.posts,
                "| Engagement:", user.engagement
            )
    def engagement_analysis(self):
        total_likes = 0
        total_comments = 0
        total_posts = len(self.posts)
        for post in self.posts:
            total_likes += post["likes"]
            total_comments += post["comments"]
        if total_posts == 0:
            avg_likes = 0
            avg_comments = 0
        else:
            avg_likes = total_likes / total_posts
            avg_comments = total_comments / total_posts
        print("\nEngagement Analytics")
        print("Total Posts:", total_posts)
        print("Total Likes:", total_likes)
        print("Total Comments:", total_comments)
        print("Average Likes per Post:", avg_likes)
        print("Average Comments per Post:", avg_comments)
    def summary_report(self):
        print("Social Media Analytics Report")
        self.trending_hashtags()
        self.most_active_users()
        self.engagement_analysis()

analyzer = SocialMediaAnalyzer()
analyzer.register_user("U101", "alex", "alex@email.com")
analyzer.register_user("U102", "john", "john@email.com")
analyzer.register_user("U103", "sara", "sara@email.com")
analyzer.add_post(
    "P101",
    "U101",
    "Learning #Python and #AI today",
    120,
    15
)

analyzer.add_post(
    "P102",
    "U102",
    "Exploring #MachineLearning and #AI trends",
    200,
    30
)

analyzer.add_post(
    "P103",
    "U103",
    "Working with #Python #DataScience",
    150,
    20
)

analyzer.add_post(
    "P104",
    "U101",
    "Building projects in #AI and #Python",
    180,
    25
)
analyzer.view_users()
analyzer.summary_report()
```
