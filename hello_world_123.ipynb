```python
trans=[1000,-500,3000,-1200,2000,-700,1500]
transactions = len(trans)
print(f"Total number of transactions: {transactions}")
print(f"first and last transaction: {trans[0]}, {trans[-1]}")
if -1200 in trans:
    trans.remove(-1200)
    print("List after removing first occurrence of -1200:", trans)
else:
    print(" -1200 not found in the list.")
trans[2]=2800
trans.sort()
trans.reverse()
print("Sorted and reversed list:", trans)

debit_count = 0
for transaction in trans:
    if transaction < 0:
        debit_count += 1
print(f"The number of debit transactions is: {debit_count}")
debit_transactions_list = [transaction for transaction in trans if transaction < 0]
print(f"List of debit transactions: {debit_transactions_list}")
print(f"Number of debit transactions: {len(debit_transactions_list)}")
```
